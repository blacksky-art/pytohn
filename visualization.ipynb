```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline

ds = pd.read_csv("car.csv")
ds.shape
```

```python
ds.isnull().sum()
```

```python
ds.plot(x = 'Engine_Size', y = 'Mileage', kind='scatter')
```

```python
ds.Brand.value_counts().plot(kind= 'bar')
```

```python
# Filter rows for 2023
ds_2023 = ds[ds['Year'] == 2023]

# Plot brand value counts for 2023
ds_2023['Brand'].value_counts().plot(kind='bar')
plt.title('Brand Distribution for 2023')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.show()
# audi has sold the most number of cars in the year 2023
```

```python
ds_2023.plot(x = 'Engine_Size', y = 'Mileage', kind='scatter')
```

```python
sns.barplot(x='Brand',y='Price', data=ds_2023, 
            hue='Transmission')
 
plt.show()
```

```python
plt.hist(ds_2023['Price'])
 
plt.title("Histogram")
 
# Adding the legends
plt.show()
```

```python
plt.hist(ds_2023['Brand'])
 
plt.title("Histogram")
 
# Adding the legends
plt.show()
```

```python
plt.hist(ds_2023['Mileage'])
 
plt.title("Histogram")
 
# Adding the legends
plt.show()
```

```python
ds['Owner_Count'].plot(kind = 'box')
```

```python
ds.plot(kind = 'box')
```

```python
ds['Price'].plot(kind = 'box')
```

```python
ds.boxplot(column='Price', by='Brand', figsize=(10, 6), rot=45)
plt.title('Price Distribution by Brand')
plt.suptitle('')  # Remove automatic subtitle
plt.xlabel('Brand')
plt.ylabel('Price')
plt.show()
'''There are outliers for price'''
```

```python
ds_2023.boxplot(column='Price', by='Brand', figsize=(10, 6), rot=45)
plt.title('Price Distribution by Brand')
plt.suptitle('')  # Remove automatic subtitle
plt.xlabel('Brand')
plt.ylabel('Price')
plt.show()
```

```python
ds_2023.plot(x = 'Engine_Size', y = 'Mileage', kind='box')
```

```python
ds_2023.boxplot(column='Mileage', by='Brand', figsize=(12, 6), rot=45)
plt.title('Mileage by Brand')
plt.suptitle('')
plt.xlabel('Brand')
plt.ylabel('Mileage')
plt.show()
```

```python
import matplotlib.pyplot as plt

ds_2023.boxplot(column='Mileage', by='Engine_Size')
plt.title('Mileage by Engine Size')
plt.suptitle('')  # optional: removes the automatic 'Boxplot grouped by' title
plt.xlabel('Engine Size')
plt.ylabel('Mileage')
plt.show()
#there are outliers in mileage for engine size 1.0 and 3.3
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))  # Set figure size before plotting
plt.pie(
    ds.Transmission.value_counts(), 
    labels=ds.Transmission.value_counts().index,  # safer than ds.Transmission.unique()
    autopct='%.2f%%'
)
plt.title('Transmission Type Distribution')
plt.show()
#distribution of cars based on transmission
```
